```python
#튜플은 어떻게 만들까?(tuple)
t1 = ()
t2 = (1, ) #★ 수정하거나 삭제할 수 없다 + 반드시 ,를 써줘야한다 ★
t3 = (1, 2, 3)
t4 = 1, 2, 3 #튜플은 괄호가 생략이 가능하다
t5 = ('a', 'b', ('ab', 'cd'))

print(type(t1))
print(type(t2))
print(type(t3))
print(type(t4))
print(type(t5))
print(t1)
print(t2)
print(t3)
print(t4)
print(t5)

```

```python
a = 1, 
print(type(a)) #하나의 요소는 int라 뜬다 만약 ,를 추가해준다면 튜플로 변한다 튜플은 한개의 요소라도 ,를 추가해줘야 한다.
```

```python
# 튜플의 요솟값을 지우거나 변경하려고 하면 어떻게 될까?

t1 = (1, 2, 'a', 'b')
#del t1[0] #튜플은 지워지는걸 허락하지 않는다 튜플은 삭제가 안된다
```

```python
# 튜플 요솟값을 변경하려 할 때
t1 = (1, 2, 'a', 'b')
#t1[0] = 'c' #튜플은 변경하는것도 허락하지 않는다. 튜플은 수정도 안된다 
```

```python
#튜플 인덱싱

t1 = (1, 2, 'a', 'b')
print(t1[2])
print(t1[-1])
```

```python
# 튜플 슬라이싱

t1 = (1, 2, 'a', 'b')
t1[1:]
```

```python
# 튜플 더하기

t1 = (1, 2, 'a', 'b')
t2 = (3, 4)
print(t1 + t2) #리스트처럼 결합이 된다
print(t1) #더하기 한다고 변경이 되진 않는다
```

```python
#튜플 곱하기

t2 = (3, 4)
print(t2 * 3)
print(t2)
```

```python
#튜플 길이 구하기

print(len(t1))
print(len(t2))
```
