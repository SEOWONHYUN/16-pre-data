```python
#문자열은 어떻게 만들고 사용할까?

## 1. 큰따옴표로 양쪽 둘러싸기
a = "Hello World"

## 2. 작은 따옴표로 양쪽 둘러싸기
b = 'Python is fun'

## 3. 큰따옴표 3개로 둘러싸기
c = """Life is too short, You need Python"""

## 4. 작은따옴표 3개로 둘러싸기
d = '''Life is too short, You need Python'''

print(a)
print(b)
print(c)
print(d)

print(type(a))
print(type(b))
print(type(c))
print(type(d))
```

```python
# 문자열 포매팅
## 1. 숫자 바로 대입

print("I eat %d apples" % 3)

## 2. 문자열 바로 대입
print("I eat %s apples" % "five") 
```

```python
## 3. 숫자값을 나타내는 변수로 대입
number = 3
print("I eat %d apples" % number)
```

```python
## 2개 이상의 값 넣기

number = 10
day = "three"
print("I eat %d apples so I was sick for %s days" % (number, day))
```

```python
# format 함수를 사용한 포매팅

print("I eat {0} apples".format(3))
print("I eat {0} apples".format("five"))
```

```python
# format함수를 이용해서 2개이상의 값 넣기
number = 10
day = "three"
print("I eat {0} apples so I was sick for {1} days".format(number, day))
```

```python
# f문자열 포매팅 

name = "홍길동"
age = 10
print(f"나의 이름은 {name}입니다. 나이는 {age}입니다.") #앞에 f 붙이면 언급한 변수들에 대한 값이 입력됨
```

```python
# 문자열 관련 함수들

a = "hobby"
print(type(a))

## 문자 개수 세기 - count
print(a.count("b")) ## b라는 글자의 갯수 세기 lan은 전체 글자의 갯수 세기
```

```python
## 위치 알려주기 - find

a = "Python is the best choice"
print(a.find("b")) ## b의 위치 찾기
print(a.find("k")) ## 만약 문자가 존재하지 않을 경우 -1 반환 있을 경우 해당하는 인덱스 반환
```

```python
# 위치 알려주기2 - index 
a = "Life is too short"
a.index("t")
#a.index("k") 없는 값은 에러가 뜬다
```

```python
# 문자열 삽입 - join 

print(",". join("abcd"))
```

```python
# 소문자를 대문자로 바꾸기 - upper

a = "hi"
a.upper()
```

```python
#대문자를 소문자로 바꾸기 - lower

a = "HI"
a.lower()
```

```python
# 공백지우기
a = "   hi   "
print(a)
a = a.strip()
print(a)
```

```python
# 공백지우기 왼쪽 / 오른쪽
a = "   hi   "
print(a.lstrip())
print(a.rstrip())
```

```python
# 문자열 바꾸기 - replace

a = "Life is too short"
a.replace("Life", "your leg")
```

```python
# 문자열 나누기 - split # 나누었을 때 결과값 확인
a = "Life is too short" #공백을 기준으로 나눔 대괄호는 리스트라 정의
a.split()
```
