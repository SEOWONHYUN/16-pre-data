```python
# 딕셔너리는 어떻게 만들까?
#{key1: value1, key2: value2, ...}

dic = {
    'name' : 'pey', 
    'phone': '010-1234-1234',
    'birth': '1118'
}
print(dic)
print(type(dic))
```

```python
# 딕셔너리 쌍 추가하기 
a = {1 : "a"}
a[2] = 'b'

print(a)
```

```python
a[3] = [1, 2, 3]
print(a)
```

```python
#딕셔너리 요소 삭제하기
print(a)
del a[1] #키값(1)만 집어넣는다 요소 전체가 삭제됨
print(a)
```

```python
# 딕셔너리에 key를 사용해서 value 얻기
grade = {
    'pey' : 10,
    'juli': 99
}

print(grade['pey'])
print(grade['juli'])
```

```python
# 딕셔너리 관련 함수
## 1. key 리스트 만들기 - keys

a = {
    'name' : 'pey', 
    'phone': '010-1234-1234',
    'birth': '1118'
}
list(a.keys())
```

```python
## value 리스트 만들기 - values

list(a.values())
```

```python
## key와 value 쌍으로 얻기 - items

print(a)
a.items()
list(a.items())
```

```python
# key와 value 쌍 모두 지우기 - clear
a.clear()
print(a)
```

```python
# key로 value 얻기 - get
a = {
    'name' : 'pey', 
    'phone': '010-1234-1234',
    'birth': '1118'
}
print(a['name'])
print(a.get('name'))
print(a.get('phone'))
```

```python
#print(a['hobby'])
#a.get('hobby', 0)---> 없는 값  뒤에 0 없는 값을 대체한다
```

```python
# 해당 key가 딕셔너리 안에 있는지 조사하기 - in
print(a)
print('name' in a)
print('email' in a)
```

```python
#key로 value 꺼내기 - pop
a = {
    'name' : 'pey', 
    'phone': '010-1234-1234',
    'birth': '1118'
}
result = a.pop("phone")
print(result)
print(a)
```

```python
# 딕셔너리의 키값으로 올 수 없는 데이터 타입은 리스트--> 변경이 가능하기 때문에 mutable
a=[1,2,3]
a[0]=3
print(a)
```

```python
a = 1 
a = 2
print(a)
```
