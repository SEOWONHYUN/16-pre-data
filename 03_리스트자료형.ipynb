```python
# 리스트는 어떻게 만들고 사용할까?

# 리스트를 만드는 형식 -> 리스트명 = [요소1, 요소2, ...]

a = [] #--> 비어있는 리스트
b = [1, 2, 3]
c = ["life", "is", "too", "short"]
d = [1, 2, "life", "is"]
e = [1, 2, ["life", "is"]]

print(type(a))
print(type(b))
print(type(c))
print(type(d))
print(type(e))
```

```python
# 리스트의 인덱싱과 슬라이싱
# 하나의 값을 가져오는건 인덱싱
a = [1, 2, 3]
print(a[0])
print(a[2])
print(type(a[2]))
```

```python
a[0] + a[2]
```

```python
a[-1]
```

```python
b = ['1', '2', '3', 'a', 'b', 'c']
b[-1]
```

```python
b[0]+b[2]+b[-1] #문자열 이므로 결합
```

```python
c = [1, 2, ['a', 'b', ['life', 'is']]]
c[-1][-1][0] #체이닝?
#len(c) #1,2,전체
```

```python
# 리스트의 슬라이싱
a = [1, 2, 3, 4, 5]

print(a[0:1])
print(a[0:2])
print(a[3:3])
```

```python
print(a[2:])
print(a[:3])
```

```python
a = [1, 2, 3, ['a', 'b', 'c'], 4, 5]

#['b', 'c']
len(a)
#print(a[-3][1:3])
```

```python
# 리스트 더하기(+)

a = [1, 2, 3]
b = [4, 5, 6]

a + b
```

```python
# 리스트 곱하기(*)

a = [1, 2, 3]
a * 3 #3번 결합 a * b는 안됨
```

```python
# 리스트의 수정과 삭제 

a = [1, 2, 3]
a[2] = 4 #3을 4로 수정하기
print(a)
```

```python
# del로 리스트 요소 삭제하기 
a = [1, 2, 3]
del a[1] #del은 삭제
print(a)
```

```python
a = [1, 2, 3, 4, 5]
del a[2:]
print(a)
```

```python
# 리스트 관련 함수
## 리스트 요소 추가하기 - append 제일 많이 쓰는 함수

a = [1, 2, 3]
a.append(4) #값을 집어넣는다.
print(a)
```

```python
a.append([5, 6])
print(a)
```

```python
# 리스트 정렬 - sort
a = [1, 4, 2, 3]
a.sort()
print(a)
a.sort(reverse=True)#뒤집어짐 관련된 함수 매서드라 함
print(a)
```

```python
a = [1, 4, 2, 3]
sorted(a)
sorted(a, reverse=True) #그냥 숫자를 뒤집는다.
```

```python
# 리스트 뒤집기 - reverse
a = ['a', 'c', 'b']
a.reverse()
print(a)
```

```python
# 인덱스 반환 - index
a = ['a', 'c', 'b']
a.index("c")
```

```python
# 리스트에 요소 삽입 - insert
a = [1, 2, 3]
a.insert(0, 4) #마지막에 집어넣을경우 append 함수 사용
print(a) 
```

```python
# 리스트 요소 제거 -remove
a = [1, 2, 3, 1, 2, 3]

a.remove(3) #앞에 있는 3이 지워진다.
a.remove(3) #한번더 3을 지워준다.
print(a)
```

```python
# 리스트 요소 꺼내기 - pop
a = [1, 2, 3]
result = a.pop() #특이하게 마지막 요소를 꺼낸다
print(result)
print(a)
```

```python
# 리스트에 포함된 요소 x의 개수 세기 - count
a = [1, 2, 3, 1]
a.count(1)
```

```python
# 리스트 확장 - extend 결합하는 형태

a = [1, 2, 3]
a.extend([4, 5])
print(a)
```
